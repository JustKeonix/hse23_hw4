```python
!for file in *.fastq;\
do\
    base=$(basename "$file" .fastq);\
    hisat2 -p 1 -x ./mm10/genome -U "$file" -S "${base}.sam" 2> "${base}.hisat";\
done
```

```python
!for file in *.sam;\
do\
    base=$(basename "$file" .sam);\
    grep -P '^@|NH:i:1$' "${base}.sam" > "${base}.uniq.sam";\
done
```

```python
!for file in *.uniq.sam;\
do\
    base=$(basename "$file" .uniq.sam);\
    echo $base;\
    grep -v '^@' "${base}.uniq.sam" | wc -l;\
done
```

```python
!for file in *.uniq.sam;\
do\
    base=$(basename "$file" .uniq.sam);\
    echo "processing ${base} file";\
    htseq-count --format=sam --stranded=no "${base}.uniq.sam"  gencode.vM25.annotation.gtf > "${base}.counts";\
done
```

```python
import pandas as pd

# File names
files = [
    "SRR3414635.counts", "SRR3414636.counts", "SRR3414637.counts",
    "SRR3414629.counts", "SRR3414630.counts", "SRR3414631.counts"
]

# Corresponding column names
column_names = ["c1", "c2", "c3", "r1", "r2", "r3"]

# Read and combine the files
dfs = []
for file, col_name in zip(files, column_names):
    df = pd.read_csv(file, sep='\t', names=['Gene', col_name], index_col=0)
    dfs.append(df)

# Merge all dataframes on the Gene index
combined_df = pd.concat(dfs, axis=1)

# Save the combined dataframe to a new file
combined_df.to_csv("ALL.counts")
```
